### Clean the text by removing punctuation and numbers since those aren't encrypted

```python
import re

cleaned = data1\
    .map(lambda x: re.sub(r'[^\D]', ' ', x))\
        .map(lambda y: re.sub(r'[^\w\s]', ' ', y))\
            .map(lambda z: re.sub('_', '', z))\
                .map(lambda word: word.strip())

cleaned.collect()
```

---

### Word and character count

```python
words1 = cleaned.flatMap(
    lambda line: line.split()
)

chars1 = cleaned.flatMap(
    lambda line: list(line)
)
```

```python
num1words = words1.count()
num1words
```

```python
num1chars = chars1.count()
num1chars
```

---

### Frequency of the words in the document

```python
words1freq = words1\
    .map(lambda word: (word, 1))\
        .reduceByKey(lambda num_word1, num_word2: num_word1+num_word2)\
            .map(lambda word: (word[0], word[1], word[1]/float(num1words)))\
                .sortBy(lambda x: x[1], ascending=False)
```

```python
words1freq.collect()
```

---

### Frequency of the characters in the document

```python
chars1freq = chars1\
    .map(lambda x: (x.lower(), 1))\
        .reduceByKey(lambda num_char1, num_char2: num_char1+num_char2)\
            .map(lambda x: (x[0], x[1], x[1]/float(num1chars)))\
                .filter(lambda x: x[0] != ' ')\
                    .sortBy(lambda x: x[1], ascending=False)
```

```python
chars1freq.collect()
```

### Assuming the most common encrypted letter decrypts to 'E'

```python
most_freq_letter = chars1freq.map(lambda x: x[0].upper()).first()
```

```python
def find_key(letter):
    return 26 + (ord('E') - ord(letter))
```

```python
encryption_key = find_key(most_freq_letter)
encryption_key
```

---

### Test the encryption key by using a random encrypted word and shifting its letters

```python
import nltk.corpus as nc

def test_key(encrypted_word, key):
    new_word = ''
    
    for letter in encrypted_word:
        shifted_letter = ord(letter) + key
        
        if shifted_letter > 90:
            shifted_letter -= 26
        new_word += chr(shifted_letter)

    if new_word.lower() in nc.words.words():
        print(new_word)
```

```python
random_word = words1.sample(withReplacement=True, fraction=0.0002)
random_word.first()
```

```python
test_key(random_word.first(), encryption_key)
```

##### it works!

```python
def decrypt(encrypted_element, key):
    new_element = ''
    for element in encrypted_element:
        shifted_element = ord(element) + key

        if shifted_element > 90:
            shifted_element -= 26
        new_element += chr(shifted_element)
        
    return new_element
```

```python
def decrypt_file(file_name):
    letters = list('ABCDEFGHIJKLMNOPQRSTUVWXYZ')
    
    with open(file_name, 'r') as f:
        txt = f.read()
    
    decrypted_txt = ''

    for char in txt:
        if char in letters:
            char = decrypt(char, encryption_key)

        decrypted_txt += char

    return decrypted_txt
```
